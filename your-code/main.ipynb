## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
# your code here
from scipy.stats import uniform

a = 10   
b = 20   

# Generate 10 random numbers between a and b
random_numbers = uniform.rvs(loc=a, scale=b - a, size=10)

print("Random numbers between 10 and 20:")
print(random_numbers)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here
def generate_uniform(bottom, ceiling, count):
    """
    Generates uniformly distributed random numbers.

    Parameters:
        bottom (float): lower boundary
        ceiling (float): upper boundary
        count (int): how many numbers to generate

    Returns:
        np.ndarray: array of random numbers
    """
    return np.random.uniform(bottom, ceiling, count)


# 2️⃣ Generate two datasets
data1 = generate_uniform(10, 15, 100)
data2 = generate_uniform(10, 60, 1000)
```

```python
# your code here
data1 = generate_uniform(10, 15, 100)
data2 = generate_uniform(10, 60, 1000)
```

```python
# your code here
plt.figure(figsize=(12, 5))
# First plot
plt.subplot(1, 2, 1)
plt.hist(data1, bins=10, color='skyblue', edgecolor='black')
plt.title('Uniform Distribution (10 to 15, n=100)')
plt.xlabel('Value')
plt.ylabel('Count')

# Second plot
plt.subplot(1, 2, 2)
plt.hist(data2, bins=10, color='lightgreen', edgecolor='black')
plt.title('Uniform Distribution (10 to 60, n=1000)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
from scipy.stats import norm 
```

```python
# your code here

def generate_normal(mean, std_dev, count):
    """
    Generates normally distributed random numbers.

    Parameters:
        mean (float): average value (μ)
        std_dev (float): standard deviation (σ)
        count (int): how many numbers to generate

    Returns:
        np.ndarray: array of random numbers
    """
    return np.random.normal(mean, std_dev, count)
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here
import numpy as np
import matplotlib.pyplot as plt


def generate_exponential(mean, size):
    """
    Generates exponentially distributed random numbers.
    
    Parameters:
      mean (float): the mean (scale) of the exponential distribution
      size (int): number of random values to generate
    """
    return np.random.exponential(scale=mean, size=size)


exp_10 = generate_exponential(mean=10, size=10)
exp_100 = generate_exponential(mean=10, size=100)


plt.figure(figsize=(12, 4))

# Histogram for size=10
plt.subplot(1, 2, 1)
plt.hist(exp_10, bins=100, color='skyblue', edgecolor='black')
plt.title('Exponential Distribution (size = 10)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Histogram for size=100
plt.subplot(1, 2, 2)
plt.hist(exp_100, bins=100, color='salmon', edgecolor='black')
plt.title('Exponential Distribution (size = 100)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your code here
import math

# Given values
mean = 10         # mean time in minutes
lmbda = 1 / mean  # λ = 1/mean
t = 15            # time in minutes

# Exponential CDF formula: P(X < t) = 1 - e^(-λt)
probability = 1 - math.exp(-lmbda * t)

print(f"Probability that a customer spends less than 15 minutes in the bank: {probability:.4f}")
```

```python
# your answer here
# Hint: This is same as saying P(x<15)
mean = 10              # mean time (minutes)
lmbda = 1 / mean       # λ = 1/mean
x = 15                 # time threshold (minutes)

# Calculate probability using the CDF: P(X < x) = 1 - e^(-λx)
P_x_less_15 = 1 - math.exp(-lmbda * x)

print(f"P(X < 15) = {P_x_less_15:.4f}")
```
